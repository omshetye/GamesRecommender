```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df["Tags"]=df["Tags"].apply(stem) 
```

text vectorization: text in tags is converted to a vector. We will use the technique called bag of words
Steps:
1. All the tags are combined to form a text containing many words
2. Few most frequently appearing words are extracted from this text 
3. Frequency of these words is checked for a word in every tag
4. The frequencies of all words in a tag is the vector of that tag
5. Tags having similar frequencies (Vectors which are closer) will be games which are similar
6. Stop-Words are not used for text vectorization (eg. in, a, for, to etc)
7. count vectorization class from scikit learn is used here

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(df['Tags']).toarray()
```

```python
vectors.shape
```

```python
vectors=vectors[0:20000, :]
vectors[0]
```

```python
#There are alot of words repeated like conflicts, conflicting, conflict
#To solve this issue we apply stemming
ps.stem("playing")
```

Now to compare games, we calculate the cosine distance between the vector of each game. euclidean distance is not reliable in higher dimensions so we use cosine. distance is inversely proportional to similarity. We use the cosine similarity function from sklearm.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectors.shape
```

```python
similarity_matrix_1=cosine_similarity(vectors[:10000, :])
similarity_matrix_2=cosine_similarity(vectors[10000:, :])
```

```python
sim_mat=np.concatenate((similarity_matrix_1, similarity_matrix_2), axis=0).astype(np.float32)
```

```python
sim_mat.shape
```

```python
sorted(list(enumerate(sim_mat[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(game):
    game_index=df[df['Title']==game].index[0]
    dist=sim_mat[game_index]
    games_list=sorted(list(enumerate(dist)), reverse=True, key=lambda x:x[1])[1:11]
    for i in games_list:
        print(df.iloc[i[0]].Title)
```

```python
recommend('Valorant')
```
